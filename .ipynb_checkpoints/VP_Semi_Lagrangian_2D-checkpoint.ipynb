```python
def make_mesh(L, LV):
    xs = linspace(0, L, nx, endpoint=False)    
    hx = xs[1]-xs[0]

    #vs = linspace(-LV, LV, nv, endpoint=False)
    vs = linspace(-LV, LV, nv, endpoint=True)
    hv = vs[1]-vs[0]

    V, X = meshgrid(vs, xs)
    return X, V, xs, vs, hx, hv

def semilag_x(deltat, f, mode = 1):
    '''
    mode 1: scipy cubic spline
    mode 2: my spline
    '''

    Xs = concatenate([xs, [xs[-1]+hx]]) # add another x point to mesh for input into cubic spline (periodic BC's)

    if mode == 1:
        for i, v in enumerate(vs):
    
            sp = CubicSpline(Xs, concatenate([f[:,i], [f[0,i]]]), bc_type='periodic')
            f[:,i] = sp(Xs[:-1] - v*deltat)
    
    elif mode == 2:
        scale_arr = np.array([-6, 2, -2, 6]) * hx**3
        for i, v in enumerate(vs):
            
            o = (-v*deltat) % hx

            #c1, c2, c3, c4 = - (o-2*hx) * (o-1*hx) * (o) / (6*hx**3), (o-2*hx) * (o-1*hx) * (o+1*hx) / (2*hx**3), -(o-2*hx) * (o) * (o+1*hx) / (2*hx**3), (o-1*hx) * (o) * (o+1*hx) / (6*hx**3)
            #kernel = [c4, c3, c2, c1]

            d0, d1, d2, d3 = o+hx, o, o-hx, o-2*hx
            kernel = np.flip(np.ones(4) * (d0*d1*d2*d3) / np.array([d0, d1, d2, d3]) / scale_arr)

            f_padded = np.concatenate([f[-1:, i], f[:, i], f[:2, i]])
            f[:,i] = np.convolve(f_padded, kernel, mode = 'valid')

            permute_index = int(v*deltat // hx) + 1

            f[:,i] = np.roll(f[:, i], permute_index)
        


def semilag_v(deltat, f, E, s=None, mode = 1):

    #Vs = concatenate([vs, [vs[-1]+hv]])

    #for i, v in zip(range(len(xs)), xs):
    if mode == 1:
        for i in range(len(xs)):
    
            #sp = CubicSpline(Vs, concatenate([f[i,:], [f[i,0]]]), bc_type='periodic')
            #f[i,:] = sp(Vs[:-1] - E[i]*deltat)
    
            sp = CubicSpline(vs, f[i,:], bc_type='not-a-knot')      
            f[i,:] = sp(vs[:] - E[i]*deltat)
            
    elif mode == 2:
        scale_arr = np.array([-6, 2, -2, 6]) * hv**3
        for i in range(len(xs)):
            o = (- E[i]*deltat) % hv

            #c1, c2, c3, c4 = - (o-2*hv) * (o-1*hv) * (o) / (6*hv**3), (o-2*hv) * (o-1*hv) * (o+1*hv) / (2*hv**3), -(o-2*hv) * (o) * (o+1*hv) / (2*hv**3), (o-1*hv) * (o) * (o+1*hv) / (6*hv**3)
            #kernel = [c4, c3, c2, c1]

            d0, d1, d2, d3 = o+hv, o, o-hv, o-2*hv
            kernel = np.flip(np.ones(4) * (d0*d1*d2*d3) / np.array([d0, d1, d2, d3]) / scale_arr)
            
            f_padded = np.concatenate([0*f[i, -1:], f[i, :], 0*f[i, :2]])
            f[i, :] = np.convolve(f_padded, kernel, mode = 'valid')

            #f[i, :] = np.convolve(f[i, :], kernel, mode = 'same') # same pads edges with 0's

            permute_index = int(E[i]*deltat // hv) + 1

            f[i, :] = np.roll(f[i, :], permute_index)
        

# # 1D Semi-Lagrangian solver #
# def semilag_x(deltat, f):
#     for i, v in zip(range(len(vs)), vs):
#         f[:,i] = interp(xs - v*deltat, xs, f[:,i], period=L)

# def semilag_v(deltat, f, E, s=None):


#     if sum(np.real(E)*deltat % hv == 0.0) != 0:
#         print("ERROR: E*deltat must _not_ be a multiple of hv")
#         exit(1)

#     for i in range(len(xs)):
#         f[i,:] = interp(vs - E[i]*deltat, vs, f[i,:], period=2.0*LV)
#         f[i,:] += deltat*s[i] if s is not None else 0.0


def compute_rho(f):
    return hv*sum(f_eq, axis=1) - hv*sum(f, axis=1)

def compute_E_from_rho(rho):
    rhohat = fft(rho)
    Ehat = 1j*zeros(len(rhohat))
    Ehat[1:] = -1.0/(1j* 2 * pi * fftfreq(nx, d=L/nx)[1:])*rhohat[1:]
    #Ehat[1:] = -1.0/(1j * fftfreq(len(rhohat), d=L/nx)[1:])*rhohat[1:]
    Ehat[0] = 0.0
    return real(ifft(Ehat))

def compute_E(f):
    return compute_E_from_rho(compute_rho(f))

def electric_energy(E):
    return 0.5*sum(E**2)*hx


def time_step_forward(dt, f, H, mode = 1):
    semilag_x(0.5 * dt, f, mode = mode)
    f_star = f.copy()
    E = compute_E(f)
    E_total = E + H  # No need to index H, since it is constant
    semilag_v(dt, f, E_total, mode = mode)
    semilag_x(0.5 * dt, f, mode = mode)
    return f_star, E_total


def run_forward(H, mode = 1):
    f = f_iv.copy()
    t = 0.0
    ees = []
    fs = [f.copy()]
    E_stars = []
    f_stars = []
    Es = []

    start_time = time.time()
    
    for i in range(num_steps):
        # current_H = H[:,i]  # No need to index H anymore        

        E = compute_E(f.copy())
        Es.append(E.copy())

        ee = electric_energy(E)
        ees.append(ee)

        f_star, E_total_star = time_step_forward(deltat, f, H, mode = mode)  # Just pass constant H
        
        fs.append(f.copy())
        f_stars.append(f_star.copy())
        E_stars.append(E_total_star)

        t += deltat

    elapsed_time = time.time() - start_time
    print('Done! Took ' + str(elapsed_time) + ' seconds for ' + str(num_steps) + ' timesteps, ' + str(elapsed_time / num_steps) + ' per timestep' )

    return fs, ees, f_stars, E_stars, Es
```
